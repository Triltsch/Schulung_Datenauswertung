### Ein erstes Beispiel

Nutzen wir NumPy, um einige Daten zu erzeugen und die matplotlib, um diese zu visualisieren:

```python
# Bibliotheken werden mit import eingebunden:
import matplotlib.pyplot as plt
import numpy as np


# Daten erzeugen
x = np.linspace(0, 10, 100) # erzeuge 100 Werte zwischen 0 und 10 für x-Achse
y = 4 + 2 * np.sin(2 * x)   # erzeuge für jeden x-Wert einen y-Wert (Funktionsgleichung)

# Plot erzeugen
fig, ax = plt.subplots()
ax.plot(x, y, linewidth=2.0, color='c') # Farben können über einfache Buchstaben angegeben werden

# einige einfache Formatierungen
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

# Plot anzeigen
plt.show()
```

Was hat das Ganze jetzt mit Arrays zu tun? <br/>
Schauen wir uns die x und y Daten einmal als Rohdaten an, dann erkennen wir das immer mehrere Werte zusammenhängend gespeichert wurden.

```python
print(x)

print(y)
```

#### Einzelne Datenpunkte mit dem Scatter-Plot visualisieren:

```python
# Daten erzeugen
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))

# Größe und Farbe der Punkte als weitere "Dimensionen"
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# Plot erzeugen
fig, ax = plt.subplots()
ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

# Plot anzeigen
plt.show()
```

#### 1. Auswertung: Eine Korrelationsanalyse: 

```python
#Note that the number 3 in the code indicates the number of digits after the decimal point for each cell.
correlation = data.corr()
correlation.style.background_gradient(cmap='coolwarm').set_precision(3)
```

Es werden hier nur die Spalten korreliert, die Zahlenwerte enthalten. Was, wenn wir den Zusammenhang der einzelnen Längen und Breiten mit der Art der Blume untersuchen wollen?<br/>
--> Wir wandeln einfach die Arten in Zahlen um: 

```python
variant = {'Versicolor': 0, 'Setosa' : 1, 'Virginica' : 2}

data['variety'] = data['variety'].map(variant)
```

Daten können auch gefiltert werden. Wir erstellen jetzt einfach für jede Variante einen einzelenn Datensatz

```python
versicolor = data[data.variety == 0]
setosa = data[data.variety == 1]
virginica = data[data.variety == 2]

setosa
```

#### 2. Auswertung: Scatterplots erzeugen
Wir wollen die sepal.length über der sepal.width darstellen
Beschreibung der Parameter: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# erst für versicolor
ax.scatter(versicolor['sepal.length'],versicolor['sepal.width'],color="red",label="versicolor",alpha= 0.5)
# jetzt für setosa
ax.scatter(setosa['sepal.length'],setosa['sepal.width'],color="green",label="setosa",alpha= 0.5)
# nun noch für virginica
ax.scatter(virginica['sepal.length'],virginica['sepal.width'],color="blue",label="virginica",alpha= 0.5)


plt.xlabel("sepal.length")
plt.ylabel("sepal.width")
plt.legend()
plt.show()
```

#### 3D-Daten plotten

Es gibt für die matplotlib ein Toolkit, mit dem sich auch 3D-Scatterplot leicht realisieren lassen.
Wir drucken beispielhaft die sepal.length auf der x-Achse, die sepal.width auf der y-Achse und die petal.width auf der z-Achse:

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

x = data['sepal.length']
y = data['sepal.width']
z = data['petal.length']

ax.scatter(x, y, z, c='r', marker='o')

ax.set_xlabel('sepal.length')
ax.set_ylabel('sepal.width')
ax.set_zlabel('petal.length')

plt.show()
```
